```python
import pandas as pd
import json
```

```python
with open('/content/drive/MyDrive/Ultimate End to End Data Analytics Project for Beginners With Cricket Dataset/t20_json_files/t20_json_files/t20_wc_match_results.json') as f:
  data = json.load(f)
data
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
#rename scorecard column as match_id
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
with open('/content/drive/MyDrive/Ultimate End to End Data Analytics Project for Beginners With Cricket Dataset/t20_json_files/t20_json_files/t20_wc_batting_summary.json') as f:
  data = json.load(f)
  all_records = []
  for rec in data:
    all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head(11)

```

```python
df_batting["out/no_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head(10)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†', ' '))
df_batting['dismissal'] = df_batting['dismissal'].apply(lambda x: x.replace('†', ' '))
df_batting.head(20)
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
  key1 = row['team1'] + ' Vs ' + row['team2']
  key2 = row['team2'] + ' Vs ' + row['team1']

  match_ids_dict[key1] = row["match_id"]
  match_ids_dict[key2] = row["match_id"]

match_ids_dict
```

```python
match_ids_dict["Namibia Vs Sri Lanka"]
```

```python
df_batting["match_id"] = df_batting['match'].map(match_ids_dict)

df_batting.head()
```

```python
df_batting.drop(columns=['dismissal'], inplace=True)
df_batting.head(10)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†', ' '))
# df_batting['dismissal'] = df_batting['dismissal'].apply(lambda x: x.replace('†', ' '))
df_batting.head(20)
```
